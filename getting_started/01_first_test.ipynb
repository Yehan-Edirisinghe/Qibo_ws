<h1>First test of Qibo Library</h1>

<p1>execution of basic functions and visualizations mostly taken from [link](https://qibo.science/qibo/stable/code-examples/examples.html)

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')    # Use dark background for plots

import qibo
assert qibo.__version__ >= "0.2.20", "This script requires qibo version 0.2.20 or higher."
from qibo import Circuit
from qibo.gates import H,CZ,M

from qibo.ui import plot_circuit, plot_density_hist, visualize_state


custom_style = {                    # Define a custom style for the circuit plot
    "facecolor" : "#000000",
    "edgecolor" : "#ffffff",
    "linecolor" : "#ffffff",
    "textcolor" : "#ffffff",
    "fillcolor" : "#c41b1b",
    "gatecolor" : "#9000ff",
    "controlcolor" : "#360000"
}
```

```python
c = Circuit(3)
c.add(H(0))
c.add(CZ(0, 1))
c.add(CZ(0,2))
c.add(M(0, 1, 2))
```

## Visualizing the Circuit

```python
plot_circuit(c, style=custom_style,scale=1.5)
plt.title("Circuit Diagram", color="white", fontsize=16)
plt.tight_layout(pad=2.0)
plt.show()
```

## Evolving the Circuit and Plotting Results

```python
init_state = np.array([1, 0, 0, 0, 0, 0, 0, 0], dtype=np.complex128)
print("Initial state:", init_state)

res = c.execute(initial_state=init_state, nshots=1000)

visualize_state(res)
plt.title("Final State Visualization", color="white", fontsize=16)
plt.tight_layout(pad=2.0)
plt.show()
```
